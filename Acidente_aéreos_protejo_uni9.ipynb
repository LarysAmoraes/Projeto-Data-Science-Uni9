## ***CENIPA disponibiliza estatística de acidentes atualizada***

Houve queda no número de acidentes aeronáuticos, em 2019. Conforme a análise dos dados feita pela Assessoria de Estatística do CENIPA, foram registrados 153 acidentes aeronáuticos, no ano passado, sendo 35 deles com fatalidade.

Quando comparados com os números de 2018 é percebida queda de 7,8% no número de acidentes e de 5,4% no número de acidentes com fatalidade. Ainda comparando os dois últimos anos da aviação civil, temos uma queda de 22,8% na quantidade de vítimas fatais.

Por segmento de aviação civil, foi registrada diminuição no número de acidentes nos segmentos “particular” e “táxi aéreo”. Ambos com queda no número de acidentes com fatalidade de 14,3% e 11,1% respectivamente.

O segmento de aviação agrícola manteve a quantidade de acidentes de um ano para o outro e, em 2019, registrou apenas um acidente com fatalidade. No segmento “regular” houve aumento: de 2 acidentes, em 2018, para 3, em 2019.

### ***Avaliando Quantidade de Ocorrencias***

```python
#Quantidade de Ocorrencias
qnt_ocorrencia = df['ocorrencia_classificacao'].value_counts()
display(qnt_ocorrencia)
```

```python
#Exibindo em Grafico Total de Ocorrencias dos Acidentes
import numpy as np 
import matplotlib.pyplot as plt

labels = ['Acidentes', 'Incidentes Graves', 'Incidentes']
vals = [954, 289, 13]


fig, ax = plt.subplots(figsize=(12, 5))

explode=(0.1, 0.1, 0)

ax.pie(vals, labels=labels, autopct='%.1f%%', shadow=True, explode=explode)
ax.set_title('Tipos de Ocorrencias', fontsize=16)
plt.show()
```

## ***Quantidade de acidentes***
